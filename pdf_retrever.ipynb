```python
!pip install pypdf
from langchain.document_loaders import PyPDFLoader

# Create the absolute path to your data directory
google_drive_mount_point = '/content/drive/MyDrive'
filename = 'yolov9_paper.pdf'
data_dir = os.path.join(google_drive_mount_point, filename)

# Load the PDF document directly using PyPDFLoader
loader = PyPDFLoader(data_dir)  # Pass the file path directly to PyPDFLoader
docs = loader.load()
```

```python
docs
```

```python
len(docs)
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=100)
texts = text_splitter.split_documents(docs)
```

```python
texts[0]
```

```python
texts[1]
```

```python
db = FAISS.from_documents(texts, HuggingFaceEmbeddings(model_name='BAAI/bge-base-en-v1.5'))
```

```python
query = "information bottleneck"
results = db.similarity_search(query)
```
